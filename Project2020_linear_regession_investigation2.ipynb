```python
#z = x
#q = y
z=df["speed"]
q=df["power"]
```

```python
np.polyfit(z,q,1)
```

```python
m,c =np.polyfit(z,q,1)
```

```python
plt.plot(z, q, 'k.',label="ScatterPlot")
#y=ax**2+bx+c, quadratic 
plt.plot(z, m * z + c, 'r-',label="Regression")
plt.xlabel("approxTurbineSpeed(ms)")
plt.ylabel("approxPowerGen(watts) ")
plt.legend(loc="best")
```

```python
a,b,c = np.polyfit(z,q,2)
```

```python
plt.plot(z, q, 'k.',label="ScatterPlot")
plt.plot(z, a * z * z + b * z + c, 'r-',label="Regression")
#plt.plot(z,a * z*z+b * z*z*+d*z+c,'r-',label="Regression")
#plt.plot(x, m * x + c, 'r-')
#plt.plot(x, m * x * x + b * x+d* z + c, 'r-')

plt.xlabel("approxTurbineSpeed(ms)")
plt.ylabel("approxPowerGen(watts) ")
plt.legend(loc="best")
```

```python
a,b,c,d = np.polyfit(z,q,3)
```

```python
a,b,c,d
```

```python
plt.plot(z, q, 'k.',label="ScatterPlot")#z = x 
#cubic polynominal y=ax**3+bx**2+cx+d;
# y = (a * x to the power of 3)+(b*x to the power of 2)+(c*x)+d
plt.plot(z, a * z **3+ b * z **2 + c*z + d,'r-',label="Regression")
plt.xlabel("approxTurbineSpeed(ms)")
plt.ylabel("approxPowerGen(watts) ")
plt.legend(loc="best")
```
